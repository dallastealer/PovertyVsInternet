```python
povertyData = pd.read_excel("drive/MyDrive/Current/DS/IncomeVsPoverty/Datasets/PovertyData.xlsx")
internetData = pd.read_excel("drive/MyDrive/Current/DS/IncomeVsPoverty/Datasets/InternetData.xlsx")
internetData = internetData[internetData["Year"] == 2019]
print(len(povertyData))
print(len(internetData))
```

```python
def tier(num):
  if num == 0:
    return 0.0
  elif num == 1:
    return 0.2
  elif num == 2:
    return 0.4
  elif num == 3:
    return 0.6
  elif num == 4:
    return 0.8
  else:
    return 1
```

```python
povertyData = povertyData[["FIPS_code", "PCTPOVALL_2020"]]
povertyData.columns = ["FIPS","PCT_POV"]
povertyData.head()
```

```python
internetData["Tier_3"] = internetData["Tier_3"] - internetData["Tier_4"]
internetData["Tier_2"] = internetData["Tier_2"] - (internetData["Tier_3"] + internetData["Tier_4"])
internetData["Tier_1"] = internetData["Tier_1"] - (internetData["Tier_2"] + internetData["Tier_3"] + internetData["Tier_4"])
```

```python
internetData["SumOfTiers"] = 1. / (internetData["Tier_1"] + internetData["Tier_2"] + internetData["Tier_3"] + internetData["Tier_4"])
internetData["PCT_200Kbps"] = ((internetData["SumOfTiers"] * internetData["Tier_1"] * internetData["Housing_Units"]) / internetData["Housing_Units"]).round(2)
internetData["PCT_10Mbps"] = ((internetData["SumOfTiers"] * internetData["Tier_2"] * internetData["Housing_Units"]) / internetData["Housing_Units"]).round(2)
internetData["PCT_25Mbps"] = ((internetData["SumOfTiers"] * internetData["Tier_3"] * internetData["Housing_Units"]) / internetData["Housing_Units"]).round(2)
internetData["PCT_100Mbps"] = ((internetData["SumOfTiers"] * internetData["Tier_4"] * internetData["Housing_Units"]) / internetData["Housing_Units"]).round(2)
internetData["Number_200Kbps"] = internetData["Housing_Units"] * internetData["PCT_200Kbps"]
internetData["Number_10Mbps"] = internetData["Housing_Units"] * internetData["PCT_10Mbps"]
internetData["Number_25Mbps"] = internetData["Housing_Units"] * internetData["PCT_25Mbps"]
internetData["Number_100Mbps"] = internetData["Housing_Units"] * internetData["PCT_100Mbps"]
```

```python
internetData = internetData[["FIPS", "State_Name", "County_Name", "Housing_Units", "PCT_200Kbps", "PCT_10Mbps", "PCT_25Mbps", "PCT_100Mbps", "Number_200Kbps", "Number_10Mbps", "Number_25Mbps", "Number_100Mbps"]]
internetData
```

```python
avgPeoplePerHousehold = 2.6 #Avg number of people per household in United States
df = pd.merge(internetData, povertyData, on="FIPS", how="left")
print(len(df))
df["Number_200Kbps"] = df["Number_200Kbps"] * avgPeoplePerHousehold
df["Number_10Mbps"] = df["Number_10Mbps"] * avgPeoplePerHousehold
df["Number_25Mbps"] = df["Number_25Mbps"] * avgPeoplePerHousehold
df["Number_100Mbps"] = df["Number_100Mbps"] * avgPeoplePerHousehold
df["Number_POV"] = (df["PCT_POV"] / 100.) * (df["Housing_Units"] / 2.6)
df
```
